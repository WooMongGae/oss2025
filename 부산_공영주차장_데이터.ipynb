```python
# 라이브러리 불러오기
import pandas as pd

# 📌 1. 엑셀 파일 불러오기
df = pd.read_excel('/content/부산광역시 공영주차장 정보 조회.xlsx')

# 📌 2. 데이터 컬럼명 확인 (KeyError 방지)
print("📌 엑셀 데이터 컬럼명 확인")
print(df.columns)  # 실제 컬럼명 출력

# 📌 3. 실제 컬럼명을 반영하여 주요 컬럼 선택
actual_columns = list(df.columns)  # 실제 컬럼명을 리스트로 변환
print("\n🔹 실제 데이터 컬럼명:", actual_columns)  # 확인용 출력

# 컬럼명을 동적으로 할당 (KeyError 방지)
col_mapping = {
    "주차장명": "주차장명",
    "소재지도로명주소": "소재지도로명주소",
    "주차구획수": "주차구획수",
    "유무료구분": "유무료구분",
    "운영구분": "운영구분"
}

# 실제 컬럼명이 다르면 자동으로 매핑
df = df.rename(columns={v: k for k, v in col_mapping.items() if v in actual_columns})

# 📌 4. 주요 컬럼 선택 (컬럼명이 실제 존재하는지 확인 후 선택)
selected_columns = [col for col in col_mapping.keys() if col in df.columns]
df_filtered = df[selected_columns]

# 📌 5. 주차 면수 기준 정렬 (내림차순)
df_sorted = df_filtered.sort_values(by='주차구획수', ascending=False)

# 📌 6. 정리된 데이터프레임 출력
print("📌 정리된 데이터")
display(df_sorted.head(10))  # 주차 면수가 많은 상위 10개 주차장 표시

# 📌 7. 운영구분별 주차장 개수 출력 (운영구분 컬럼이 존재할 경우)
if "운영구분" in df.columns:
    print("\n🚗 운영구분별 공영주차장 개수")
    print(df["운영구분"].value_counts())
else:
    print("⚠️ '운영구분' 컬럼이 데이터에 없습니다. 운영구분별 개수 분석을 생략합니다.")
```
